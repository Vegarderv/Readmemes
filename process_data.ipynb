```python
import json

with open("temp.json", "r", encoding="utf8") as f:
    loaded = json.loads(f.read())

loaded
```

```python
users = {}

for user in loaded:
    try:
        users[user["id"]] = user["real_name"]
    except:
        users[user["id"]] = user["name"]

with open("users.json", "w", encoding="utf8") as f:
    f.write(json.dumps(users, ensure_ascii=False))
```
